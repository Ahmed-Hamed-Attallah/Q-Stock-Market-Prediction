#### ✅ Check for Stationarity

```python
# Create a new series that is the difference between each day's close and the previous day's close
stock['close_diff'] = stock['close'].diff()

# Drop the first row which will be NaN after differencing

close_diff = stock['close_diff'].dropna()
# ADF test on this new differenced series.
result = adfuller(close_diff)

print(f'ADF Test for Differenced Close Prices')
print(f'ADF Statistic: {result[0]:.4f}')
print(f'p-value: {result[1]:.4f}')
print('Critical Values:')
for key, value in result[4].items():
    print(f'   {key}: {value:.4f}')

if result[1] <= 0.05:
    print("Result: Stationary (reject null hypothesis)")
else:
    print("Result: Non-stationary (fail to reject null hypothesis)")
print('-'*50)

# Create subplots to compare original and differenced series
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))

# Original series
ax1.plot(df['close'])
ax1.set_title('Original Close Prices')
ax1.set_ylabel('Price')
ax1.grid(True)

# Differenced series
ax2.plot(close_diff)
ax2.axhline(y=0, color='r', linestyle='-', alpha=0.7)
ax2.set_title('First-Order Differenced Series (Stationary)')
ax2.set_ylabel('Price Difference')
ax2.set_xlabel('Time')
ax2.grid(True)

plt.tight_layout()
plt.show()
```

### ✅ Model Building

#### Preparaing data 

```python
stock = stock.sort_index()

# 1. Enhanced Feature Engineering
df = stock.copy()

# Technical Indicators with error handling
# Simple Moving Averages
df['sma_7'] = ta.sma(df['close'], length=7)
df['sma_21'] = ta.sma(df['close'], length=21)
df['ema_12'] = ta.ema(df['close'], length=12)
df['ema_26'] = ta.ema(df['close'], length=26)

# RSI
df['rsi_14'] = ta.rsi(df['close'], length=14)

# MACD with error handling
macd_data = ta.macd(df['close'])
if macd_data is not None:
    macd_cols = [col for col in macd_data.columns if 'MACD' in col]
    if len(macd_cols) >= 2:
        df['macd'] = macd_data[macd_cols[0]]
        df['macd_signal'] = macd_data[macd_cols[1]]

# Stochastic with error handling
stoch_data = ta.stoch(df['high'], df['low'], df['close'])
if stoch_data is not None:
    stoch_cols = [col for col in stoch_data.columns if 'STOCH' in col]
    if len(stoch_cols) >= 2:
        df['stoch_k'] = stoch_data[stoch_cols[0]]
        df['stoch_d'] = stoch_data[stoch_cols[1]]

# Bollinger Bands with correct column names
bb_data = ta.bbands(df['close'], length=20)
if bb_data is not None:
    # Get the actual column names from the returned DataFrame
    bb_columns = bb_data.columns
    if len(bb_columns) >= 3:
        df['bb_upper'] = bb_data[bb_columns[0]]  # First column is upper band
        df['bb_lower'] = bb_data[bb_columns[1]]  # Second column is lower band  
        df['bb_middle'] = bb_data[bb_columns[2]]  # Third column is middle band

# ATR
df['atr'] = ta.atr(df['high'], df['low'], df['close'], length=14)

# Volume indicators
df['volume_sma'] = ta.sma(df['volume'], length=10)
df['volume_ratio'] = df['volume'] / df['volume_sma']
df['obv'] = ta.obv(df['close'], df['volume'])

# Price-based features
df['price_range'] = (df['high'] - df['low']) / df['close']
df['price_change'] = df['close'].pct_change()
df['high_low_ratio'] = df['high'] / df['low']

# Lag features
for lag in [1, 2, 3, 5, 10]:
    df[f'close_lag_{lag}'] = df['close'].shift(lag)
    df[f'volume_lag_{lag}'] = df['volume'].shift(lag)

# Rolling statistics
df['close_roll_mean_7'] = df['close'].rolling(7).mean()
df['close_roll_std_7'] = df['close'].rolling(7).std()
df['close_roll_mean_21'] = df['close'].rolling(21).mean()

# Target variable
df['target'] = df['close'].shift(-1)

# Drop NaN values
df_clean = df.dropna()

print("Enhanced data shape:", df_clean.shape)
print("Columns:", df_clean.columns.tolist())

# 2. Prepare features and target
feature_columns = [col for col in df_clean.columns if col != 'target']
X_all = df_clean[feature_columns].values
y_all = df_clean[['target']].values

# 3. Enhanced Scaling
price_cols = ['close', 'open', 'high', 'low', 'sma_7', 'sma_21', 'ema_12', 'ema_26']
price_cols += [col for col in df_clean.columns if 'close_lag' in col or 'roll_mean' in col or 'bb_' in col]

volume_cols = ['volume', 'volume_sma', 'volume_ratio'] + [col for col in df_clean.columns if 'volume_lag' in col]

indicator_cols = ['rsi_14', 'macd', 'macd_signal', 'stoch_k', 'stoch_d', 'atr', 'obv', 
                 'price_range', 'price_change', 'high_low_ratio', 'close_roll_std_7']

# Filter only columns that exist in our dataframe
price_cols = [col for col in price_cols if col in feature_columns]
volume_cols = [col for col in volume_cols if col in feature_columns]
indicator_cols = [col for col in indicator_cols if col in feature_columns]

print(f"Price columns: {len(price_cols)}")
print(f"Volume columns: {len(volume_cols)}")
print(f"Indicator columns: {len(indicator_cols)}")

# Initialize scalers
scaler_price = MinMaxScaler(feature_range=(0, 1))
scaler_volume = StandardScaler()
scaler_indicator = MinMaxScaler(feature_range=(-1, 1))
scaler_target = MinMaxScaler(feature_range=(0, 1))

# Scale features
X_scaled = np.zeros((len(df_clean), len(feature_columns)))

# Scale price columns
if price_cols:
    price_indices = [feature_columns.index(col) for col in price_cols]
    X_scaled[:, price_indices] = scaler_price.fit_transform(df_clean[price_cols])

# Scale volume columns  
if volume_cols:
    volume_indices = [feature_columns.index(col) for col in volume_cols]
    X_scaled[:, volume_indices] = scaler_volume.fit_transform(df_clean[volume_cols])

# Scale indicator columns
if indicator_cols:
    indicator_indices = [feature_columns.index(col) for col in indicator_cols]
    X_scaled[:, indicator_indices] = scaler_indicator.fit_transform(df_clean[indicator_cols])

# Scale target
y_scaled = scaler_target.fit_transform(y_all)

print("Scaling completed")

# 4. Create Sequences
SEQ_LENGTH = 60
PRED_LENGTH = 1

Xs, ys = [], []
for i in range(len(X_scaled) - SEQ_LENGTH - PRED_LENGTH + 1):
    Xs.append(X_scaled[i:i+SEQ_LENGTH])
    ys.append(y_scaled[i+SEQ_LENGTH:i+SEQ_LENGTH+PRED_LENGTH])

X = np.array(Xs)
y = np.array(ys)

print("X shape:", X.shape)
print("y shape:", y.shape)
```

#### ✅ Train and Test Split (80/20)

```python
# 5. Train/Test Split
split_idx = int(len(X) * 0.8)
X_train, X_test = X[:split_idx], X[split_idx:]
y_train, y_test = y[:split_idx], y[split_idx:]

print("Train shapes:", X_train.shape, y_train.shape)
print("Test shapes:", X_test.shape, y_test.shape)
```

#### ✅ Build LSTM model

```python
# Build LSTM model
model = Sequential([
    LSTM(128, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2]), recurrent_dropout=0.2),
    BatchNormalization(),
    Dropout(0.3),
    
    LSTM(64, return_sequences=True, recurrent_dropout=0.2),
    BatchNormalization(),
    Dropout(0.3),
    
    LSTM(32, return_sequences=False, recurrent_dropout=0.1),
    BatchNormalization(),
    Dropout(0.2),
    
    Dense(64, activation='relu'),
    BatchNormalization(),
    Dropout(0.2),
    
    Dense(32, activation='relu'),
    Dropout(0.1),
    
    Dense(16, activation='relu'),
    Dense(1)
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='mse',
    metrics=['mae', 'mape']
)

model.summary()
```

#### ✅ Train

```python
callbacks = [
    EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-7, verbose=1),
    ModelCheckpoint('best_lstm_model.keras', monitor='val_loss', save_best_only=True, verbose=1)
]

history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=100,
    batch_size=32,
    callbacks=callbacks,
    verbose=1,
    shuffle=False
)

# Load best model
model.load_weights('best_lstm_model.keras')
```

#### ✅ Evaluation 

```python
# 8. Make Predictions
y_pred_scaled = model.predict(X_test)

# Inverse transform
y_pred_actual = scaler_target.inverse_transform(y_pred_scaled)
y_test_actual = scaler_target.inverse_transform(y_test.reshape(-1, 1))

# 9. Calculate Metrics
mse = mean_squared_error(y_test_actual, y_pred_actual)
mae = mean_absolute_error(y_test_actual, y_pred_actual)
rmse = np.sqrt(mse)
r2 = r2_score(y_test_actual, y_pred_actual)
mape = np.mean(np.abs((y_test_actual - y_pred_actual) / y_test_actual)) * 100

print("\n" + "="*50)
print("EVALUATION RESULTS")
print("="*50)
print(f"MAE  : {mae:.4f}")
print(f"MSE  : {mse:.4f}")
print(f"RMSE : {rmse:.4f}")
print(f"R²   : {r2:.4f}")
print(f"MAPE : {mape:.2f}%")

# 10. Plot Results
plt.figure(figsize=(15, 10))

# Plot 1: Training history
plt.subplot(2, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss History')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()

# Plot 2: Actual vs Predicted
plt.subplot(2, 2, 2)
plt.plot(y_test_actual, label='Actual Price', alpha=0.7)
plt.plot(y_pred_actual, label='Predicted Price', alpha=0.7)
plt.title('Actual vs Predicted Prices')
plt.xlabel('Time Steps')
plt.ylabel('Price')
plt.legend()

# Plot 3: Scatter plot
plt.subplot(2, 2, 3)
plt.scatter(y_test_actual, y_pred_actual, alpha=0.5)
plt.plot([y_test_actual.min(), y_test_actual.max()], 
         [y_test_actual.min(), y_test_actual.max()], 'r--', lw=2)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Predicted vs Actual Correlation')

# Plot 4: Residuals
plt.subplot(2, 2, 4)
residuals = y_test_actual - y_pred_actual
plt.plot(residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.title('Prediction Residuals')
plt.xlabel('Time Steps')
plt.ylabel('Residuals')

plt.tight_layout()
plt.show()

# 11. Feature Importance Analysis
print("\nTop 10 Most Important Features:")
feature_correlations = []
for i, col in enumerate(feature_columns):
    if i < X.shape[2]:
        try:
            corr = np.corrcoef(X_test[:, :, i].flatten(), y_test.flatten())[0, 1]
            feature_correlations.append((col, abs(corr)))
        except:
            continue

feature_correlations.sort(key=lambda x: x[1], reverse=True)
for feature, corr in feature_correlations[:10]:
    print(f"{feature:20} | Correlation: {corr:.4f}")

# Save final model
model.save("final_lstm_model.keras")
print("\nModel training completed and saved!")
```
